## Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

## Formatos

```python
# Para mostrar todas las columnas
pd.set_option('display.max_columns', None)

# Para que no salga warning al graficar
mpl.rcParams['figure.max_open_warning'] = 0
```

## Funciones

```python
def plot_variable(df, col, desc):
    
    """
    Genera un gráfico de línea para una columna específica de un DataFrame con formato personalizado.

    Parámetros:
    -----------
    df : pandas.DataFrame
        DataFrame que contiene los datos.
    col : str
        Nombre de la columna del DataFrame que se va a graficar.
    desc : str
        Descripción o título secundario que se mostrará debajo del nombre de la columna.
        
    """
    
    # Se inicia la firgura
    fig, ax = plt.subplots(figsize=(10, 4))
    
    # Plot
    ax.plot(df[col], lw=2)

    ### Formatos
    # Letra mas grande
    ax.tick_params(labelsize=14)

    # Titulo: Nombre y descripcion
    ax.text(0.05, 1.15, col, transform=ax.transAxes, fontsize=18, color='darkblue')
    ax.text(0.05, 1.05, desc, transform=ax.transAxes, fontsize=12, color='gray')

    # Se quitan los margenes de arriba y a la derecha
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Se agregan lineas horizontales
    ax.yaxis.grid(True, which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
```
