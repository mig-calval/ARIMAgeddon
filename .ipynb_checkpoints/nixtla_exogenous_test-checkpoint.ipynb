```python
import lightgbm as lgb
import pandas as pd
from mlforecast import MLForecast
from mlforecast.lag_transforms import ExpandingMean, RollingMean
from mlforecast.utils import generate_daily_series, generate_prices_for_series
```

```python
series = generate_daily_series(
    100, equal_ends=True, n_static_features=2
).rename(columns={'static_1': 'product_id'})
series.head()
```

```python
dates = sorted(series['ds'].unique())
```

```python
last_dates = dates[-7:]
```

```python
train = series[series['ds'] < last_dates[0]]
```

```python
test = series[series['ds'] >= last_dates[0]]
```

```python
prices_catalog = generate_prices_for_series(train, horizon=7)
prices_catalog.head()
```

```python
train['unique_id'].value_counts()
```

```python
train_with_prices = train.merge(prices_catalog, how='left')
test_with_prices = test.merge(prices_catalog, how='left')
```

```python
from mlforecast.utils import PredictionIntervals
```

```python
fcst = MLForecast(
    models=lgb.LGBMRegressor(n_jobs=1, random_state=0, verbosity=-1),
    freq='D',
    lags=[7],
    lag_transforms={
        1: [ExpandingMean()],
        7: [RollingMean(window_size=14)],
    },
    date_features=['dayofweek', 'month'],
    num_threads=2,
)
fcst.fit(train_with_prices, static_features=['static_0', 'product_id'],  
         prediction_intervals=PredictionIntervals(n_windows=3, h=7),)
```

```python
fcst.ts.features_order_
```

```python
levels = [50, 80, 95]
# forecasts = mlf.predict(48, level=levels)
```

```python
preds = fcst.predict(h=7, X_df=prices_catalog, level=levels)
preds.head()
```
