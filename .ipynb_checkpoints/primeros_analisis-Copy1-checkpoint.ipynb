## Funciones

```python
def plot_formatos(ax):
    
    """
    Aplica formato estético estándar a un eje de Matplotlib.

    Parámetros:
    -----------
    ax : matplotlib.axes.Axes
        Objeto de ejes sobre el que se aplicarán los formatos.        
    """
    
    # Letra mas grande
    ax.tick_params(labelsize=14)
    
    # Se quitan los margenes de arriba y a la derecha
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Se agregan lineas horizontales
    ax.yaxis.grid(True, which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
```

```python
def plot_variable(df, col, desc):
    
    """
    Genera un gráfico de línea para una columna específica de un DataFrame con formato personalizado.

    Parámetros:
    -----------
    df : pandas.DataFrame
        DataFrame que contiene los datos.
    col : str
        Nombre de la columna del DataFrame que se va a graficar.
    desc : str
        Descripción o título secundario que se mostrará debajo del nombre de la columna.
        
    """
    
    # Se inicia la firgura
    fig, ax = plt.subplots(figsize=(10, 4))
    
    # Plot
    ax.plot(df[col], lw=2, color='#1f77b4')    

    # Titulo: Nombre y descripcion
    ax.text(0.05, 1.15, col, transform=ax.transAxes, fontsize=18, color='#0a4f7f')
    ax.text(0.05, 1.05, desc, transform=ax.transAxes, fontsize=12, color='gray')

    # Formatos
    plot_formatos(ax)
```

```python
def plot_dos_variables(df, col1, col2, cor=None):
    
    """
    Genera un gráfico de línea para dos columnas específicas de un DataFrame con formato personalizado.

    Parámetros:
    -----------
    df : pandas.DataFrame
        DataFrame que contiene los datos.
    col1 : str
        Nombre de la primera columna del DataFrame que se va a graficar.
    col2 : str
        Nombre de la segunda columna del DataFrame que se va a graficar.
    cor : float
        Valor de la correlación entre ambas columnas.
    """
    
    # Se inicia la firgura
    fig, ax = plt.subplots(figsize=(10, 4))
    
    # Plot
    ax.plot(df[col1], lw=2, color='#1f77b4')
    ax.plot(df[col2], lw=2, color='indianred')

    # Titulo: Nombre y descripcion
    ax.text(0.05, 1.15, col1, transform=ax.transAxes, fontsize=18, color='#0a4f7f')
    ax.text(0.05, 1.05, col2, transform=ax.transAxes, fontsize=18, color='indianred')

    if cor is not None:
        ax.text(0.75, 1.1, f'Corr. = {round(100*cor, 2)}%', transform=ax.transAxes, fontsize=18, color='gray')
        
    # Formatos
    plot_formatos(ax)
```

```python
# Convertir a formato largo
corr_long = corr.reset_index().melt(id_vars='index')
corr_long.columns = ['col1', 'col2', 'corr']

# Quitar duplicados y autocorrelaciones
corr_long = corr_long[corr_long['col1'] != corr_long['col2']]
corr_long['pair'] = corr_long.apply(lambda row: tuple(sorted([row['col1'], row['col2']])), axis=1)
corr_long = corr_long.drop_duplicates(subset='pair').drop(columns='pair')

# Absoluto
corr_long['corr_abs'] = np.abs(corr_long['corr'])
```

```python
corr_long.sort_values(by='corr_abs', ascending=False).head(20)
```
